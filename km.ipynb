Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading Dataset

```python
ds = pd.read_csv('Mall_Customers.csv')
x = ds.iloc[:, 3:].values
```

Finding No. of Clusters by Elbow Method

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, init= 'k-means++', random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

Model Training

```python
km = KMeans(n_clusters=5, init='k-means++', random_state=42)
km.fit(x)
y_km = km.predict(x)
```

Visualizing the Model

```python
plt.scatter(x[y_km==0, 0], x[y_km==0, 1], c='red', s=100, label='cluster1')
plt.scatter(x[y_km==1, 0], x[y_km==1, 1], c='blue', s=100, label='cluster1')
plt.scatter(x[y_km==2, 0], x[y_km==2, 1], c='green', s=100, label='cluster1')
plt.scatter(x[y_km==3, 0], x[y_km==3, 1], c='yellow', s=100, label='cluster1')
plt.scatter(x[y_km==4, 0], x[y_km==4, 1], c='pink', s=100, label='cluster1')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 150, c='grey', label='centroid')

plt.title('Mall Customers')
plt.xlabel('Salary')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
